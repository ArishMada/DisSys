```python
def same_month(date1, date2):
    datetime1 = datetime.strptime(date1, "%m/%d/%Y")
    datetime2 = datetime.strptime(date2, "%m/%d/%Y")

    return datetime1.year == datetime2.year and datetime1.month == datetime2.month
```

```python
def get_average():
    des18 = 0
    jan19 = 0
    feb19 = 0
    mar19 = 0
    apr19 = 0
    may19 = 0
    jun19 = 0
    july19 = 0
    aug19 = 0
    sep19 = 0
    oct19 = 0
    nov19 = 0
    dec19 = 0
    n1 = 0
    n2 = 0
    n3 = 0
    n4 = 0
    n5 = 0
    n6 = 0
    n7 = 0
    n8 = 0
    n9 = 0
    n10 = 0
    n11 = 0
    n12 = 0
    n13 = 0
    
    avg = {}
    
    for index, row in show_data.iterrows():
        date = row["Date"]
        quantity = row["Quantity"]
        
        if same_month(date, "12/01/2018"):
            des18 += quantity
            n1 += 1
        elif same_month(date, "01/01/2019"):
            jan19 += quantity
            n2 += 1
        elif same_month(date, "02/01/2019"):
            feb19 += quantity
            n3 += 1
        elif same_month(date, "03/01/2019"):
            mar19 += quantity
            n4 += 1
        elif same_month(date, "04/01/2019"):
            apr19 += quantity
            n5 += 1
        elif same_month(date, "05/01/2019"):
            may19 += quantity
            n6 += 1
        elif same_month(date, "06/01/2019"):
            jun19 += quantity
            n7 += 1
        elif same_month(date, "07/01/2019"):
            july19 += quantity
            n8 += 1
        elif same_month(date, "08/01/2019"):
            aug19 += quantity
            n9 += 1
        elif same_month(date, "09/01/2019"):
            sep19 += quantity
            n10 += 1
        elif same_month(date, "10/01/2019"):
            oct19 += quantity
            n11 += 1
        elif same_month(date, "11/01/2019"):
            nov19 += quantity
            n12 += 1
        elif same_month(date, "12/01/2019"):
            dec19 += quantity
            n13 += 1
            
            
    avg["December 2018"] = round(des18 / n1, 2)
    avg["January 2019"] = round(jan19 / n2, 2)
    avg["February 2019"] = round(feb19 / n3, 2)
    avg["March 2019"] = round(mar19 / n4, 2)
    avg["April 2019"] = round(apr19 / n5, 2)
    avg["May 2019"] = round(may19 / n6, 2)
    avg["June 2019"] = round(jun19 / n7, 2)
    avg["July 2019"] = round(july19 / n8, 2)
    avg["Aug 2019"] = round(aug19 / n9, 2)
    avg["September 2019"] = round(sep19 / n10, 2)
    avg["October 2019"] = round(oct19 / n11, 2)
    avg["November 2019"] = round(nov19 / n12, 2)
    avg["December 2019"] = round(dec19 / n13, 2)
    
    return avg

avg = get_average()
```

```python
def fix_avg(avg):
    average = {}
    month = []
    average1 = []
    
    for item, key in avg.items():
        month.append(item)
        average1.append(key)
    
    average["Month"] = month
    average["Average"] = average1
    
    return average

avg_month = fix_avg(avg)

avg_df = pd.DataFrame(avg_month)
avg_df.to_csv("avg.csv", index = False)
```

```python
def get_quantity(price):
    sales = {}
    inventory_data = {}
    max_inventory_range = 2000
    
    for index, row in show_data.iterrows():
        product_name = row["ProductName"]
        quantity = row["Quantity"]
        
        if product_name in sales:
            sales[product_name] += quantity
        else:
            sales[product_name] = quantity

    for index, row in show_data.iterrows():
        product_name = row["ProductName"]
        sales_quantity = row["Quantity"]
        
        inventory_quantity = random.randint(1, max_inventory_range)
        
        for item, key in sales.items():
            if product_name == item:
                inventory_data[product_name] = [key, inventory_quantity]
                
        for item, key in price.items():
            if product_name == item:
                quantity1, inv_quantity = inventory_data[product_name]
                inventory_data[product_name] = [quantity1, inv_quantity, key]
        
    return sales, inventory_data

sales, inventory_data = get_quantity(prices)
inventory_data
```

```python
def fix_dict(inventory_data):
    inventory = {}
    product = []
    sales = []
    inventories = []
    prices = []
    
    for item, key in inventory_data.items():
        product.append(item)
        sales.append(key[0])
        inventories.append(key[1])
        prices.append(key[2])
        
    
    inventory["ProductName"] = product
    inventory["Sales"] = sales
    inventory["Inventory"] = inventories
    inventory["Price"] = prices
    
    return inventory

inventory = fix_dict(inventory_data)

inventory_df = pd.DataFrame(inventory)
inventory_df.to_csv("inventory.csv", index = False)
```

```python
def get_highest_quantity():
    sales = {}
    highest_product = None
    highest_quantity = 0

    for index, row in show_data.iterrows():
        product_name = row["ProductName"]
        quantity = row["Quantity"]

        if product_name in sales:
            sales[product_name] += quantity
        else:
            sales[product_name] = quantity

        if quantity > highest_quantity:
            highest_quantity = quantity
            highest_product = product_name

    return highest_product, highest_quantity

highest_product, highest_quantity = get_highest_quantity()
print(f"The product with the highest quantity is {highest_product} with a quantity of {highest_quantity}")
```
